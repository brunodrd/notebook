# <b><font color="#9356CA">DEFINITION :</b>

Un **<font color="#9356CA">labyrinthe parfait**</font> est un labyrinthe dans lequel toutes les cases peuvent être atteintes, et cela avec un seul chemin possible.

Un tel labyrinthe, peut donc être représenté par un arbre (son graphe ne comporte pas de cycles)

**1.a Ce labyrinthe est il parfait ?**


**1.b Pourquoi l'arbre correspondant est-il binaire ?**






**2. Faire la représentation de ce labyrinthe avec  un graphe (sur feuille).**

Conventions utilisées :  

1.   Les cases sont numérotées colones,ligne. Par exemple l'entrée est 04 et la sortie 44.
2.   Lorsqu'un noeud (une case) n'a qu'un fils, **on considérera que c'est un fils gauche,
 même si on tourne à droite**
3.   De même, si à une intersection on peut aller tout droit ou à droite, on considèrera que tout droit est à gauche.


**3.  Quelle est la hauteur de l'arbre ?**



**4. Compréhension de l'arbre :**

4a. Dans ce labyrinthe, que peut-on dire des feuilles ?

4b. Et que dire d'un noeud ayant seulement un fils gauche ?

<font color=lightseagreen size=6><b>Implémentation d'arbres avec 2 classes</b></font>

Voici les fonctions de l'interface dont vous aurez besoin :


```
# Notre arbre sera modélisé avec un dictionnaire.
arbre = Arbre(etiquette) :
    Crée un objet arbre, dont la racine porte l'étiquette passée en paramètre
    Attributs :
        arbre.noeuds : dictionnaire dont les clés sont les etiquettes des sommets et les valeurs les objets noeuds associés
        arbre.etiquette_racine : etiquette de la racine de l'arbre

############### Primitives d'accès  ###############

arbre.empty():
    return True si Arbre est un arbre vide, False sinon

arbre.filsD(etiquette) :
    renvoie True si le noeud dont l'etiquette est donnée en paramètre à un fils droit, False sinon.
        
arbre.filsG(etiquette) :
    renvoie True si le noeud dont l'etiquette est donnée en paramètre à un fils droit, False sinon.

arbre.racine() :
    renvoie l'objet noeud de la racine
        Les objets Noeuds disposent de 3 attributs :
            fg et fd : objets Noeud fils gauche et droit
            val : etiquette du noeud fils

############### Primitives d'ajouts ###############

arbre.addG(etiquette_parent, etiquette_fils) -> None :
    :param etiquette_parent: est une etiquette d'un noeud de l'arbre
    :param etiquette_fils: est l'etiquette du fils gauche à ajouter 

    addG ajoute un fils gauche au noeud parent
        


arbre.addD(etiquette_parent, etiquette_fils) -> None :
    :param etiquette_parent: est une etiquette d'un noeud de l'arbre
    :param etiquette_fils: est l'etiquette du fils droit à ajouter 

    addD ajoute un fils droit au noeud parent


############### Primitives d'extraction de sous arbres ###############

arbre.sad(etiquette=None):
    si le paramètre cle est absent, la méthode renvoie le SAD de la racine
    si une etiquette et mentionnée, la méthode renvoie le SAD du noeud portant cette etiquette

arbre.sag(etiquette=None):
    si le paramètre cle est absent, la méthode renvoie le SAG de la racine
    si une etiquette et mentionnée, la méthode renvoie le SAG du noeud portant cette etiquette

############### Primitive de représentation ###############

print(arbre) : permet d'afficher l'arbre en mode texte

arbre.show(etiquette=None,chemin = None):
    Dessin de l'arbre.
    Si etiquette est None, la fonction dessinne l'arbre complet
    Sinon elle dessine le sous arbre dont la racine est etiquette
    Le paramètre chemin peut$etre ajouté :c'est une liste d'étiquette. Tous les noeuds de cette liste seront représentés en vert.
```
